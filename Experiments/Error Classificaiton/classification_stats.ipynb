```
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```
from google.colab import drive

drive.mount("/content/gdrive")
```

```
ERR_LBLS = ["CO", "GM", "LX", "OT"]
```

```
primaryPath = "/content/gdrive/My Drive/NLP/Translations/xlm/final/random/"
```

```
# output = {"head4":{"70000":{'CO': 0, 'GM': 0, 'LX': 0, 'OT': 0}}}

output={}
```

```
excluded_buds = []
```

##Multiple layers

```
layers = os.listdir(primaryPath)
for t in tqdm(range(len(layers))):
  layer = layers[t]
  print("Layer:", layer)
  if layer.startswith("."):
    continue
  layerPath = primaryPath + "/" + layer
  translations = os.listdir(layerPath)

  for i in range(len(translations)):
    translation = translations[i]
    # print("Translation folder:", translation)
    if(translation.startswith(".")):
      continue
    layer = translation.split("_")[2]
    # layer = translation.split("_")[2]+"_"+translation.split("_")[3]   # use this for best/ folder
    # layer = translation
    print("\n\tLayer:", layer)
    subdir = os.listdir(layerPath + "/" + translation)
    
    sub = subdir[0]
    if(subdir[0].startswith(".")):
      sub = subdir[1]
    budget = sub
    print("\tbudget:", budget)
    if budget in excluded_buds:
      continue
    # print("\tSubdirectory:", sub)
    error_file_path = layerPath + "/" + translation + "/" + sub + "/hypotheses/" + "error_predictions.csv"
    # print("\t\tFile:", error_file_path)
    df = pd.read_csv(error_file_path)
    # print(df)
    err_dict = {}
    for err in ERR_LBLS:
      res = (df[err].values == True).sum()
      err_dict[err] = res
    if(layer not in output.keys()):
      output[layer] = {}
    print("\tlayer, budget", layer, budget)
    output[layer][budget] = err_dict
  
```

##single layer 

```
translations = os.listdir(primaryPath)

for i in tqdm(range(len(translations))):
  translation = translations[i]
  print("Translation folder:", translation)
  if(translation.startswith(".")):
    continue
  layer = translation.split("_")[2]
  layer = layer.strip("budget")
  # layer = translation
  # print("\tLayer:", layer)
  subdir = os.listdir(primaryPath + "/" + translation)
  
  sub = subdir[0]
  if(subdir[0].startswith(".")):
    sub = subdir[1]
  budget = sub
  if budget in excluded_buds:
    continue
  print("\tSubdirectory:", sub)
  error_file_path = primaryPath + "/" + translation + "/" + sub + "/hypotheses/" + "error_predictions.csv"
  print("\t\tFile:", error_file_path)
  df = pd.read_csv(error_file_path)
  
  err_dict = {}
  for err in ERR_LBLS:
    res = (df[err].values == True).sum()
    err_dict[err] = res
  if(layer not in output.keys()):
    output[layer] = {}
  print("layer, budget", layer, budget)
  
  output[layer] = err_dict
 
```
